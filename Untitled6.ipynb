```python

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
data = {
    'Weather': [1, 3, 2, 4, 1, 2, 3, 4],
    'Speed': [60, 80, 50, 100, 70, 65, 90, 85],
    'Vehicle_Type': [1, 2, 1, 2, 1, 1, 2, 2],
    'Road_Type': [1, 2, 3, 1, 2, 3, 1, 2],
    'Accident_Severity': [3, 7, 4, 9, 5, 4, 8, 7]
}

df = pd.DataFrame(data)
X = df[['Weather', 'Speed', 'Vehicle_Type', 'Road_Type']]
y = df['Accident_Severity']
model = LinearRegression()
model.fit(X, y)
print("Model Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
new_data = pd.DataFrame({
    'Weather': [2],
    'Speed': [75],
    'Vehicle_Type': [1],
    'Road_Type': [2]
})

prediction = model.predict(new_data)
print("Predicted Accident Severity:", prediction[0])
plt.scatter(df['Speed'], y, color='blue')
plt.plot(df['Speed'], model.predict(X), color='red')
plt.xlabel('Speed')
plt.ylabel('Accident Severity')
plt.title('Accident Severity vs Speed')
plt.show()
```
