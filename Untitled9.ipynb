```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
a = 1664525
c = 1013904223
m = 2**32
seed = 12345
n = 10000

# LCG
numbers = []
x = seed
for i in range(n):
    x = (a * x + c) % m
    numbers.append(x / m)

numbers = np.array(numbers)

# Histogram
plt.hist(numbers, bins=30, edgecolor='black')
plt.title("Histogram of LCG Random Numbers")
plt.show()

# Runs Test
median = np.median(numbers)
runs = 1
for i in range(1, len(numbers)):
    if (numbers[i] > median) != (numbers[i-1] > median):
        runs += 1

n1 = sum(numbers > median)
n2 = sum(numbers <= median)

expected_runs = ((2*n1*n2)/(n1+n2)) + 1
variance_runs = (2*n1*n2*(2*n1*n2 - n1 - n2)) / (((n1+n2)**2)*(n1+n2-1))
z = (runs - expected_runs)/np.sqrt(variance_runs)

print("Runs Test Z-value:", z)
```

```python
import numpy as np

np.random.seed(42)

arrival_rate = 1/3
simulation_time = 300  # 5 hours in minutes

arrival_times = []
time = 0
while time < simulation_time:
    time += np.random.exponential(3)
    arrival_times.append(time)

service_times = np.random.uniform(1,4,len(arrival_times))

start_times = []
finish_times = []
waiting_times = []

for i in range(len(arrival_times)):
    if i == 0:
        start = arrival_times[i]
    else:
        start = max(arrival_times[i], finish_times[i-1])

    finish = start + service_times[i]

    start_times.append(start)
    finish_times.append(finish)
    waiting_times.append(start - arrival_times[i])

avg_wait = np.mean(waiting_times)
max_queue = max(np.cumsum(np.array(arrival_times) < simulation_time))
utilization = sum(service_times) / simulation_time
prob_wait_5 = np.mean(np.array(waiting_times) > 5)

print("Average Waiting Time:", avg_wait)
print("Server Utilization:", utilization)
print("Probability Wait > 5:", prob_wait_5)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

arrival_rate = 40
service_time = 0.015
simulation_time = 10
buffer_limit = 100

arrival_times = np.cumsum(np.random.exponential(1/arrival_rate, 1000))
arrival_times = arrival_times[arrival_times < simulation_time]

queue = []
buffer_sizes = []
current_time = 0
departures = 0
loss = 0

for arrival in arrival_times:
    while queue and queue[0] <= arrival:
        queue.pop(0)
        departures += 1

    if len(queue) < buffer_limit:
        service_start = max(arrival, queue[-1] if queue else arrival)
        queue.append(service_start + service_time)
    else:
        loss += 1

    buffer_sizes.append(len(queue))

print("Average Buffer Occupancy:", np.mean(buffer_sizes))
print("Packet Loss Probability:", loss/len(arrival_times))
print("Throughput:", departures/simulation_time)

plt.plot(buffer_sizes)
plt.title("Buffer Size Over Time")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

k = 0.08
T_env = 23
T0 = 38
dt = 0.1
time = np.arange(0,60,dt)

T = np.zeros(len(time))
T[0] = T0

for i in range(1,len(time)):
    T[i] = T[i-1] - k*(T[i-1] - T_env)*dt

plt.plot(time,T)
plt.title("Temperature vs Time")
plt.xlabel("Time (minutes)")
plt.ylabel("Temperature (C)")
plt.show()

# Find when T = 25
index = np.where(T <= 25)[0][0]
print("Time when T reaches 25C:", time[index])
```
